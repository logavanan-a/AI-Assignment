### PYTHON PROGRAMMING
Welcome to assignment on Python Programming. In this assignment you are supposed to process/read/work on a file name “Employee.csv”. Which contain employee information of an organization.Following information’s are available for each employee:

•	EID – Employee ID

•	Employee Name

•	Gender

•	Salary

•	Date of Joining

```python
import csv
from datetime import datetime
from collections import defaultdict
```

```python
def read_employee_data(filename="Employee.csv"):
    employees = []
    with open(filename, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            employees.append(row)
    return employees
```

#### In this assignment you are supposed to write functions for following objective:
•	Given an employee id, greet them with their first name.

o	Input

	E123D

o	Output

	“Hello, Mr Nitin”


```python
def greet_employee(emp_id, employees):
    for emp in employees:
        if emp["EID"] == emp_id:
            name = emp["Employee Name"].split()[0]  # first name
            if emp["Gender"].lower().startswith('m'):
                return f"Hello, Mr {name}"
            else:
                return f"Hello, Ms {name}"
    return "Employee not found."
```

#### •	Function for calculating the average salary of each department.

```python
def average_salary_by_department(employees):
    dept_salary = defaultdict(list)
    for emp in employees:
        dept = emp["Department"]
        salary = float(emp["Salary"])
        dept_salary[dept].append(salary)

    avg_salary = {dept: sum(sals)/len(sals) for dept, sals in dept_salary.items()}
    return avg_salary
```

#### •	Given an Employee id, display respective employees Name, Data of Joining and Department.
o	Input

	E123D

o	Output

	Nitin Bhole | 01-11-2016 | Products


```python
def get_employee_details(emp_id, employees):
    for emp in employees:
        if emp["EID"] == emp_id:
            name = emp["Employee Name"]
            doj = emp["Date of Joining"]
            dept = emp["Department"]
            return f"{name} | {doj} | {dept}"
    return "Employee not found."
```

#### •	Function for sorting employees from Oldest to Newest.

```python
def sort_employees_by_joining_date(employees):
    def parse_date(date_str):
        return datetime.strptime(date_str, "%d-%m-%Y")
    
    sorted_employees = sorted(employees, key=lambda x: parse_date(x["Date of Joining"]))
    return sorted_employees
```

```python
if __name__ == "__main__":
    employees = read_employee_data("Employee.csv")

    print(greet_employee("E123D", employees))

    print("\nAverage Salary by Department:")
    for dept, avg in average_salary_by_department(employees).items():
        print(f"{dept}: {avg:.2f}")

    print("\nEmployee Details:")
    print(get_employee_details("E123D", employees))

    print("\nEmployees sorted by Date of Joining:")
    for emp in sort_employees_by_joining_date(employees):
        print(f"{emp['Employee Name']} - {emp['Date of Joining']}")
```
